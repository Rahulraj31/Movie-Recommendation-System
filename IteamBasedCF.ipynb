# Item - Based Collaborative Filtering
Let us start by importing the movie lense dataset

```python
import pandas as pd
import numpy as np
```

```python
r_cols=['user_id','movie_id','rating']
ratings =pd.read_csv('ml-100k/u.data',sep='\t',names=r_cols, 
                   usecols=range(3))  #selects 3 columns with index 0 1 2 
m_cols = ['movie_id','title']
movies=pd.read_csv('ml-100k/u.item',sep='|',names=m_cols, 
                   usecols=range(2))
ratings=pd.merge(movies,ratings)
```

```python
ratings.head()
```

Now we will make pivot table 

```python
userRating=ratings.pivot_table(index=['user_id'],columns='title',values='rating')
userRating.head()
```

```python
corrMatrix= userRating.corr()
corrMatrix.head()
```

```python
corrMatrix= userRating.corr(method="pearson", min_periods=100)
#finds corr between movies rated by 100 people min.
corrMatrix.head()
```

correlation of a movie with themselfes is also NaN if they are rated by less than 100 people

```python
myRatings = userRating.iloc[10,:30].dropna()
myRatings
```

```python
simCandidates=pd.Series()
for i in range(0,len(myRatings.index)):
    print(f"Adding sims for {myRatings.index[i]}-")
    
    # retreive similar movies to this one that i rated 
    sims= corrMatrix[myRatings.index[i]].dropna()
    
    #now Scale its similarity by how weill i liked it` 
    sims=sims.map(lambda x:x*myRatings[i]) #multiply our rating with genral ratings of similar movied
    
    #Add the scores to the list  of similarity candidates
    simCandidates =simCandidates.append(sims)
    
print("Sorting..........")
simCandidates.sort_values(inplace=True, ascending = False)
print(simCandidates.head(10))
```

```python
''''
To merge duplicate entires present  in this variable produced by each film in myRatings 
as they were similar to more than one movies. this makes recommendation more strong
'''
simCandidates =simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace=True, ascending = False)
print(simCandidates.head(10))
```
